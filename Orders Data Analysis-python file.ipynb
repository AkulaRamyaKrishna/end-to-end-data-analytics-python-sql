#### --> Reload the dataset and treat specific strings as missing values (NaN)

```python
df = pd.read_csv('orders.csv',na_values = ['Not Available', 'unknown',])
df['Ship Mode'].unique()
```

#### -->  Convert all column names to lowercase for consistency
#### -->  This helps avoid case-sensitive issues during analysis

```python
df.columns = df.columns.str.lower()
df.columns
```

#### --> Replace spaces in column names with underscores
#### --> This makes column names easier to use in Python and SQL queries

```python
df.columns = df.columns.str.replace(" ","_")
df.columns
```

```python
df.head()
```

#### --> Feature Engineering â€“ discount , sale price and profit

```python
df['discount'] = df['list_price'] * df['discount_percent'] / 100
df.head()
```

```python
df['sale_price'] = df['list_price'] - df['discount']
df.head()
```

```python
df['profit'] = df['sale_price'] - df['cost_price']
df.head()
```

#### --> Convert order_date column from string to datetime format

```python
df['order_date'] = pd.to_datetime(df['order_date'],format = '%Y-%m-%d')
```
